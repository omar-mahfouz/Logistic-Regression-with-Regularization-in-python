```python
import numpy as np

class logistiRegression :    

    def __init__(self, alpha=0.01,lam=0.1 ):
               
        self.alpha = alpha        
        
        self.lam = lam
        
        self.IsTrain = False
        
        
        
    def __sigmoid(self,Num) :
        
         return  1.0 / (1.0 + np.exp(-1.0 * Num))
        
    
    
    
    def __CostFunc(self,Predictions,Y,m) :

        cost = np.sum((-1*Y*np.log(Predictions))-((1-Y)*(np.log(1-Predictions))),axis=0) / m             
        
        cost += 0.5 * self.lam * ( np.sum(self.theta * self.theta ,axis=0) - self.theta[0]*self.theta[0] ) / m 
    
        return cost
    
    
    

    
    def Predict(self,X) :   
        
        if self.IsTrain == False :
            return False
               
        # Fix DataSet
        X = np.append(1, X)

        Predictions = np.dot(X, self.theta)
        
        Predictions = self.__sigmoid(Predictions)
        
        Predictions = np.round(Predictions)
        
        return Predictions
    
    
    
    
    def __CalclutePredictions(self,X) :
        
        Predictions = np.dot(X, self.theta)
        
        Predictions = self.__sigmoid(Predictions)
        
        return Predictions
                
        

    def __UpdateWeight(self,Predictions,X,Y,n,m) :
                                    
        Error = Predictions - Y
        
        Gradient = np.dot(X.T, Error)

        self.theta[0] -= self.alpha * (1/m) * Gradient[0]
    
        for i in range(1,n) :
            self.theta[i] = self.theta[i] * (1-self.lam*self.alpha/m) - self.alpha * (1/m) * Gradient[i]
        
        return True  
    
    
        
        
    def Train(self,X,Y,iters) :
        
        m , n = X.shape        
        
        #Initiate The Weights
        self.theta = np.zeros(n+1)
        
        # Fix DataSet
        z = np.ones((n,1))
        X = np.append(z, X, axis=1)

        
        for It in range(0,iters) :
            
            # Claclute The Predictions Value  for all Training Set
            Predictions = self.__CalclutePredictions(X)

            # Calclute New Cost
            newcost = self.__CostFunc(Predictions,Y,m)

            # Calclute Derivative and update Theta
            self.__UpdateWeight(Predictions,X,Y,n,m)            
           
        self.IsTrain = True 
        
        return True
```
